```python
class NegativeBinomialDistribution:
    def __init__(self,r,p):
        self.r = r
        self.p = p
    @staticmethod
    def combination(n,r):
        return math.factorial(n)/(math.factorial(n-r)*math.factorial(r))
    def negativeBinomial(self,x):
        return self.combination(x+self.r-1,x)*self.p**self.r*(1-self.p)**x
    
    def negativeBinomialDistribution(self,*args):
        return sum([self.negativeBinomial(i) for i in args])
    
    def plot(self, maxVal):
        x = range(maxVal+1)
        y = [self.negativeBinomial(i) for i in x]
        plt.plot(x,y)
        plt.xlabel("number of failure before success")
        plt.ylabel("Probability of failure")
    
```

```python
nbd = NegativeBinomialDistribution(6,0.6)
```

```python
nbd.plot(10)
```
