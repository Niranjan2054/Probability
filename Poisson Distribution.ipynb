```python
class PoissonDistribution:
    def __init__(self,mean):
        self.mean = mean
    def poisson(self,x):
        return (math.exp(-self.mean) * self.mean**x) / math.factorial(x)
    def PoissonDistribution(self,*args):
        return sum([self.poisson(i) for i in args])
    def Graph(self):
        x = range(0,20)
        y = [self.poisson(i) for i in x]
        df = pd.DataFrame({'x':x,'probability':y})
        print(df)
        plt.plot(x,y)
    def Bar(self):
        x = range(0,20)
        y = [self.poisson(i) for i in x]
        df = pd.DataFrame({'x':x,'probability':y})
        print(df)
        plt.bar(x,y)
```

Average number of traffic accident in a locality is found to be 2.75 per week. Find the probability that during neext week, there would be
1. no accident & no. of week with no accident
2. Exactly two accident
3. less than & equal to two accident
4. At least two accident

```python
p = PoissonDistribution(2.75)
```

```python
p.Graph()
```

```python
p.Bar()
```

```python
# no accident and no. of week with no accidennt
print("Probabiliy that it will have no accident:",p.PoissonDistribution(0))
print("No. of week with no accident:",int(p.PoissonDistribution(0)*52))
```

```python
#exactly two accident and no. of week with two accident
print("Probabiliy that it will have no accident:",p.PoissonDistribution(2))
print("No. of week with no accident:",int(p.PoissonDistribution(2)*52))
```

```python
#less than and equal to two accident
print("Probability that there would be lessthan equal to two accident:",p.PoissonDistribution(0,1,2))
```

```python
#At least two accident
print("Probability that there would be atleast two accident:",1-p.PoissonDistribution(0,1))
```
