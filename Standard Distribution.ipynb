# Mathematical problem of Normal Distribution

In an entrance exam of IOE, there were 12000 students.Their mark follow a Normal Distribution with mean 64 and S.D = 13
1. Find the number of students with score more than 85
2. Probability that score lies within 60 to 75
3. Percentage of Student with score less than 50

```python
import math
class NormalDistribution:
    '''
        It is used to calculate the Normal Distribution
        By default it assign the mean =0 and variance =1 
        and Hence, it calculate the Standard Distribution
    '''
    def __init__(self,mean=0,sd=1):
        self.mean=mean
        self.sd=sd

    def distributionFunction(self,x):
        return 1/(self.sd*math.sqrt(2*math.pi))*math.exp(-1/2*((x-self.mean)/self.sd)**2)
    
    def NormalDistribution(self,lowerlimit,upperlimit):
        sub_interval = 3000
        step_size = (upperlimit-lowerlimit)/sub_interval
        integration = self.distributionFunction(lowerlimit) + self.distributionFunction(upperlimit)
        for i in range(1,sub_interval):
            k=lowerlimit+i*step_size
            if i%3==0:
                integration= integration +2 * self.distributionFunction(k)
            else:
                integration= integration +3 * self.distributionFunction(k)
        integration = integration*step_size*3/8
        return integration 
```

```python
N = 12000
mean = 64
Sd = 13
problem = NormalDistribution(mean,Sd)
scoremorethan85 = problem.NormalDistribution(85,10000) * N #here 10000 acts as infinity
```

```python
scoremorethan85
```

Hence, 637 students out of 12000 has score more than 85

```python
#probability that score lies between 60 to 75
sixtyto75 = problem.NormalDistribution(60,75)
print(sixtyto75)
```

Hence, 0.42210830052555887 is probability that score lies between 60 to 75

```python
#percentage of Student with score less than 50
scorelessthan50 = problem.NormalDistribution(-1000,50)*100 # -1000 acts as negative infinity
print(scorelessthan50,'%')
```

Hence, 14.07% of student has score less than 50

```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
class NormalDistribution:
    '''
        It is used to calculate the Normal Distribution
        By default it assign the mean =0 and variance =1 
        and Hence, it calculate the Standard Distribution
    '''
    def __init__(self,mean=0,sd=1):
        self.mean=mean
        self.sd=sd

    def distributionFunction(self,x):
        return 1/(self.sd*math.sqrt(2*math.pi))*math.exp(-1/2*((x-self.mean)/self.sd)**2)
    
    def NormalDistribution(self,lowerlimit,upperlimit):
        sub_interval = 3000
        step_size = (upperlimit-lowerlimit)/sub_interval
        integration = self.distributionFunction(lowerlimit) + self.distributionFunction(upperlimit)
        for i in range(1,sub_interval):
            k=lowerlimit+i*step_size
            if i%3==0:
                integration= integration +2 * self.distributionFunction(k)
            else:
                integration= integration +3 * self.distributionFunction(k)
        integration = integration*step_size*3/8
        return integration 
    def Graph(self):
        x = np.arange(self.mean-5*self.sd,self.mean+5*self.sd,0.01)
        y = [self.distributionFunction(i) for i in x]
        plt.plot(x,y)
```
